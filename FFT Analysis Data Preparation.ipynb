Importing the modules

```python
import pandas as pd #reading csv, creating dataframe
import matplotlib.pyplot as plt #visualising all the data
import numpy as np #array
from scipy.fftpack import rfft, rfftfreq, ifft #fourier transformation package
plt.style.use("seaborn")
#adding comments
```

```python
b25_1 = pd.read_csv("p_wave/Before/25/B25_1.csv")
b200_1 = pd.read_csv("p_wave/Before/200/B200_1.csv")
a200_1 = pd.read_csv("p_wave/After/200/A200_1.csv")
b400_A = pd.read_csv("p_wave/Before/400/B400_A.csv")
a400_A = pd.read_csv("p_wave/After/400/A400_A.csv")
```

Matplotlib plot sizes

```python
w_size = (16,6) #waveform plot size
f_size = (16,8) #frequency plot size
```

```python
time = np.arange(start=0, stop=1000, step=0.5)
```

```python
plt.figure(figsize=w_size)
plt.plot(time, b25_1.iloc[18].astype(float))
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (%)")
plt.title("B25 Sample 1 Waveform")
```

```python
plt.figure(figsize=w_size)
plt.plot(time, b200_1.iloc[18].astype(float))
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (%)")
plt.title("B200 Sample 1 Waveform")
```

```python
plt.figure(figsize=w_size)
plt.plot(time, a200_1.iloc[18].astype(float))
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (%)")
plt.title("A200 Sample 1 Waveform")
```

```python
ft_b200_1 = rfft(b200_1.iloc[18][50:950].astype(float))
freq_b200_1 = rfftfreq(len(b200_1.iloc[18][50:950]), 1/54000)
plt.figure(figsize=f_size)
plt.plot(freq_b200_1, np.abs(ft_b200_1))
plt.title("B200 Sample 1 Frequency Series")
```

```python
ft_a200_1 = rfft(a200_1.iloc[18][50:950].astype(float))
plt.figure(figsize=f_size)
plt.plot(np.abs(ft_a200_1))
plt.title("A200 Sample 1 Frequency Series")
```

```python
plt.figure(figsize=(16,6))
plt.plot(time, b400_A.iloc[18].astype(float))
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (%)")
plt.title("B400 Sample A Waveform")
```

```python
plt.figure(figsize=(16,8))
ft_b400_A = rfft(b400_A.iloc[18][50:950].astype(float))
freq_b400_A = rfftfreq(len(b400_A.iloc[18][50:950]))
plt.plot(freq_b400_A, np.abs(ft_b400_A))
plt.title("B400 Sample A Frequency Series")
```

```python
plt.figure(figsize=(16,6))
plt.plot(time, a400_A.iloc[18].astype(float))
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (%)")
plt.title("A400 Sample A Waveform")
```

```python
ft_a400_A = rfft(a400_A.iloc[18][50:950].astype(float))
freq_a400_A = rfftfreq(len(a400_A.iloc[18][50:950]), 1/54000)
plt.figure(figsize=f_size)
plt.plot(freq_a400_A, np.abs(ft_a400_A))
plt.title("A400 Sample A Frequency Series")
```
